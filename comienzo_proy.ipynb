Mi primer proyecto se va a basar en poder sacar la información que deseemos a traves de los datos extraídos del buscador 'Google'.

```python
# Importamos las librerías que vamos a necesitar

import pandas as pd
import numpy as np
```

### A continuación vamos a establecer la variable con todos los datos, para después convertirlo a dataframe y así poder realizar las consultas que queramos"

```python
data = {'Búsqueda': ['Despacho abogados', 'Abogado Madrid', 'Abogado Alcalá', '¿Cuánto cuesta un abogado?', 'Abogado Torrejón', 'Gabinete jurídico', 'Concurso acreedores', 'Asesoría', 'Última oportunidad', 'Juzgado Alcalá'],
        'Nacional': [700, 1800, 200, 700, 200, 500, 150, 2000, 500, 100],
        'Comunidad de Madrid': [300, 1200, 150, 400, 150, 350, 100, 1400, 350, 80],
        'Corredor del Henares': [50, 200, 50, 80, 50, 70, 30, 200, 70, 30]}

df = pd.DataFrame(data)


df

```

## A continuación vamos a incluir más búsquedas relacionadas con el sector jurídico

```python

# Definir las búsquedas como índice
busquedas = ["Juzgado de Madrid", "Derecho Penal", "Deuda judicial", "Análisis viabilidad", 
             "Entidad Financiera", "Deuda personal", "Préstamo Personal", "Préstamo Madrid", "Préstamos"]
             
# Crear el dataframe con las columnas correspondientes
df1 = pd.DataFrame(columns=["Nacional", "Comunidad de Madrid", "Corredor del Henares"], index=busquedas)

# Ingresar los valores en el dataframe
df1.loc["Juzgado de Madrid"] = [150, 100, 40]
df1.loc["Derecho Penal"] = [600, 400, 50]
df1.loc["Deuda judicial"] = [50, 30, 10]
df1.loc["Análisis viabilidad"] = [20, 10, 5]
df1.loc["Entidad Financiera"] = [1500, 1000, 100]
df1.loc["Deuda personal"] = [400, 300, 50]
df1.loc["Préstamo Personal"] = [600, 400, 100]
df1.loc["Préstamo Madrid"] = [150, 100, 50]
df1.loc["Préstamos"] = [3500, 2500, 300]

# Mostrar el dataframe
df1
```

### Vamos a juntar este DataFrame con el anterior que ya teníamos llamado busq

```python
combo=pd.concat([busq, df1])
combo
```

Vamos a ver si los valores máximos y mínimos han cambiado.


```python
combo.max()
```

```python
combo.min()
```

### Veamos aquellas búsquedas que tienen la palabra Madrid y vamos a sacar los sus valores máximos correspondientes

```python
combo.index
```

```python
combo_madrid=combo[combo.index.str.contains("Madrid")]
combo_madrid
```

```python
# Aquí sacamos los valos máximos de cada búsqueda
combo_madrid.max()
```

```python
# En este caso hemos convertido el tipo de número a float para que idxmax lo detectase y pudiese darnos que búsqueda es la que tiene el mayor valor de "Nacional"

max_index = combo_madrid["Nacional"].astype("float64").idxmax()
max_index

```
